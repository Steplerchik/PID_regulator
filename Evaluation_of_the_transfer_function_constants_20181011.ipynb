```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from mpmath import mp
```

#### S creation for Laplace domain

```python
tmax=20
ds = 1. / 120
smax = 1 / (0.1) / 10 *20
s = np.linspace(0.1, smax, num=int((smax-0.1) / ds))
```

#### Data load

```python
t_x = np.array(np.load('torque_times.npy'))
t_y = np.array(np.load('ang_vel_times.npy'))

xt = np.array(np.load('torque_fyaw.npy'))
yt = np.array(np.load('ang_vel_fyaw.npy'))
```

#### Changes of Data for further numerical integration

```python
start_time = t_x[0]
t_x = t_x - start_time
t_y = t_y - start_time

t_y = np.append([0],t_y)
yt = np.append([0],yt)

t_y_1 = np.delete(t_y,[len(t_y) - 1])
t_y_2 = np.delete(t_y,[0])
dty = t_y_2 - t_y_1
yt = np.delete(yt,[len(yt) - 1])
t_y = np.delete(t_y,[len(t_y) - 1])

t_x_1 = np.delete(t_x,[len(t_x) - 1])
t_x_2 = np.delete(t_x,[0])
dtx = t_x_2 - t_x_1
xt = np.delete(xt,[len(xt) - 1])
t_x = np.delete(t_x,[len(t_x) - 1])
```

#### Direct Laplace transformation of data, evaluation of the transfer function

```python
# xs = np.sum(xt[:, np.newaxis] * np.exp(-s * t_x[:, np.newaxis]) * dtx[:, np.newaxis] , axis = 0)

# ys = np.sum(yt[:, np.newaxis] * np.exp(-s* t_y[:, np.newaxis]) * dty[:, np.newaxis], axis = 0)

xs = np.sum(xt[:, np.newaxis] * np.exp(-s * t_x[:, np.newaxis]), axis = 0)

ys = np.sum(yt[:, np.newaxis] * np.exp(-s* t_y[:, np.newaxis]), axis = 0)

xt_offset = xt - np.sum(xt * dtx) / np.sum(dtx)

xs_offset = np.sum(xt_offset[:, np.newaxis] * np.exp(-s * t_x[:, np.newaxis]) * dtx[:, np.newaxis] , axis = 0)

hs = ys / xs
```

#### Time domain Plot

```python
plt.plot(t_x, xt_offset, label = u'xt_off (torq)')
plt.plot(t_x, xt, label = u'xt (torq)')
plt.plot(t_y, yt, label = u'yt (ang vel)')

plt.grid(True)
plt.xlabel(u'time')
plt.ylabel(u'f(t)')
plt.title(u'Time domain')
plt.legend()
plt.show()
```

#### Laplace domain Plot

```python
plt.plot(s, xs_offset, label = u'xs_off (torq)')
plt.plot(s, xs, label = u'xs (torq)')
plt.plot(s, ys, label = u'ys (ang vel)')

plt.grid(True)
plt.xlabel(u's')
plt.ylabel(u'f(s)')
plt.title(u'Laplace domain')
plt.legend()
plt.show()

#plt.plot(s[10:], hs[10:])
```

#### Approximation with the Least squares method, evaliation of transfer function constants

```python
def gen_data(s_, k, tet, t2):
    y = k * np.exp(-tet * s_) / s_ / (t2 * s_ + 1)
    return y

def fun(x, s_, hs_):
    return x[0] * np.exp(-x[1] * s_) / s_ / (x[2] * s_ + 1) - hs_

x0 = np.ones(3)

from scipy.optimize import least_squares

res_lsq = least_squares(fun, x0, args=(s, hs))

s_test = s
hs_lsq = gen_data(s_test, *res_lsq.x)

plt.plot(s, hs, 'o')
plt.plot(s_test, hs_lsq, 'k', linewidth=2, label='lsq')

plt.grid(True)
plt.show()

coef = res_lsq.x
print(coef)
```

```python
def gen_data(s_, k, tet, t1, t2):
    y = k * np.exp(-tet * s_) / (t1 * s_ + 1)/ (t2 * s_ + 1)
    return y

def fun(x, s_, hs_):
    return x[0] * np.exp(-x[1] * s_) / (x[2] * s_ + 1) / (x[3] * s_ + 1) - hs_

x0 = np.ones(4)

from scipy.optimize import least_squares

res_lsq = least_squares(fun, x0, args=(s, hs))

s_test = s
hs_lsq = gen_data(s_test, *res_lsq.x)

plt.plot(s, hs, 'o')
plt.plot(s_test, hs_lsq, 'k', linewidth=2, label='lsq')

plt.grid(True)
plt.show()

coef = res_lsq.x
print(coef)
```
